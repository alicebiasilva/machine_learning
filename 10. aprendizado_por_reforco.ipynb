# Aprendizado por refor√ßo

Aprendizagem por Refor√ßo (ou Reinforcement Learning - RL) √© um campo da intelig√™ncia artificial (IA) onde um agente aprende a tomar decis√µes interagindo com um ambiente com o objetivo de maximizar uma fun√ß√£o de recompensa ao longo do tempo.

√â uma forma de aprendizado onde o agente aprende por tentativa e erro, recebendo feedback (recompensas ou penalidades) com base nas a√ß√µes que toma, e ajusta suas decis√µes futuras com base nesse feedback.

A aprendizagem por refor√ßo √© formalizada atrav√©s de um processo de Markov conhecido como MDP (Markov Decision Process). Ele envolve os seguintes elementos:

* Agente: A "intelig√™ncia" que toma decis√µes (por exemplo, um rob√¥, um jogador de xadrez, ou um sistema de recomenda√ß√£o).
* Ambiente: O mundo em que o agente interage e onde ele tenta alcan√ßar um objetivo (por exemplo, um tabuleiro de xadrez ou o sistema que responde √†s a√ß√µes do usu√°rio).
* A√ß√£o: As decis√µes que o agente pode tomar no ambiente (por exemplo, mover uma pe√ßa no xadrez ou recomendar um filme).
* Estado: A configura√ß√£o do ambiente ap√≥s a a√ß√£o do agente. O estado √© a "situa√ß√£o" do sistema.
* Recompensa: O feedback recebido ap√≥s tomar uma a√ß√£o em um determinado estado. A recompensa pode ser positiva (quando o agente realiza uma boa a√ß√£o) ou negativa (quando o agente comete um erro).
* Pol√≠tica: A estrat√©gia ou plano do agente para decidir qual a√ß√£o tomar em cada estado. Pode ser determin√≠stica ou probabil√≠stica.
* Valor: A expectativa da recompensa futura que o agente espera receber a partir de um estado espec√≠fico, dependendo das a√ß√µes que ele toma.

O objetivo principal do agente √© maximizar a recompensa acumulada ao longo do tempo. O agente n√£o sabe inicialmente quais a√ß√µes levar√£o a boas recompensas, ent√£o ele explora o ambiente, tenta a√ß√µes diferentes e aprende com os resultados dessas a√ß√µes.

A recompensa em RL √© normalmente diferida ‚Äî ou seja, o agente pode tomar uma a√ß√£o e n√£o saber imediatamente se foi boa ou ruim, mas sim em etapas posteriores. Isso torna a aprendizagem por refor√ßo uma t√©cnica poderosa, mas tamb√©m complexa.

Na aprendizagem por refor√ßo, os m√©todos podem ser classificados em tr√™s grandes categorias com base na abordagem que utilizam para aprender e tomar decis√µes: m√©todos baseados em valor, m√©todos baseados em pol√≠ticas e m√©todos baseados em modelos. Cada um desses m√©todos lida com o problema de otimiza√ß√£o do comportamento do agente de maneira diferente. Vamos entender cada um desses m√©todos em detalhes:
* M√©todos baseados em valor (como Q-learning) estimam valores de estados ou a√ß√µes e usam essas estimativas para melhorar a pol√≠tica.
* M√©todos baseados em pol√≠ticas (como Policy Gradient) aprendem diretamente a pol√≠tica do agente, ajustando a distribui√ß√£o de a√ß√µes para maximizar a recompensa.
* M√©todos baseados em modelos constroem ou utilizam um modelo do ambiente para simular transi√ß√µes e recompensas e usar esse conhecimento para planejar a√ß√µes de maneira eficiente.

**Alguns dos principais m√©todos e algoritmos em aprendizagem por refor√ßo incluem:**
1. Q-Learning: Um algoritmo off-policy que tenta aprender o valor de a√ß√£o de cada par estado-a√ß√£o. Ele usa uma tabela chamada Q-table para armazenar esses valores e atualiza-os com base na recompensa recebida.
2. Deep Q-Learning (DQN): Uma extens√£o do Q-Learning que usa redes neurais profundas para aproximar a fun√ß√£o de valor Q, permitindo que o agente lide com espa√ßos de estado muito grandes.
3. Policy Gradient Methods: M√©todos que tentam aprender uma pol√≠tica direta (em vez de uma fun√ß√£o de valor). Em vez de usar uma Q-table, o agente aprende uma fun√ß√£o de probabilidade para escolher a√ß√µes, otimizando essa pol√≠tica com o tempo.
4. Actor-Critic Methods: Combina elementos dos m√©todos de value-based (como Q-Learning) e policy-based (como os m√©todos de gradiente de pol√≠tica), com dois componentes principais: o ator (que decide a a√ß√£o) e o cr√≠tico (que avalia a a√ß√£o).
5. Proximal Policy Optimization (PPO): Um m√©todo mais recente, baseado em pol√≠tica, que fornece uma maneira eficiente e est√°vel de otimizar a pol√≠tica de forma cont√≠nua.

5. Defina a pol√≠tica gananciosa

Lembre-se de que temos duas pol√≠ticas, pois o Q-Learning √© um algoritmo off-policy. Isso significa que estamos usando uma pol√≠tica diferente para atuar e atualizar a fun√ß√£o de valor.

* Pol√≠tica Epsilon-gananciosa (pol√≠tica de atua√ß√£o)
* Greedy-policy (pol√≠tica de atualiza√ß√£o)

A pol√≠tica gananciosa tamb√©m ser√° a pol√≠tica final que teremos quando o agente Q-learning for treinado. A pol√≠tica gulosa √© usada para selecionar uma a√ß√£o da tabela Q.



```python
def greedy_policy(Qtable, state):
  # Exploitation
  action = np.argmax(Qtable[state][:])

  return action
```

6. Defina a pol√≠tica gananciosa de epsilon 

Epsilon-greedy √© a pol√≠tica de treinamento que lida com a troca de explora√ß√£o/explora√ß√£o.

A ideia com epsilon-greedy:

* Com probabilidade 1‚Ää-‚Ää…õ : fazemos exploitation (ou seja, nosso agente seleciona a a√ß√£o com o maior valor do par estado-a√ß√£o).
* Com probabilidade …õ: fazemos exploration (tentativa de a√ß√£o aleat√≥ria).

E √† medida que o treinamento avan√ßa, progressivamente reduzimos o valor do epsilon, pois precisaremos de cada vez menos exploration e mais exploitation.



```python
def epsilon_greedy_policy(Qtable, state, epsilon):
  # Gera aleatoriamente um n√∫mero entre 0 e 1
  random_int = random.uniform(0,1)
  # if random_int > maior que epsilon --> exploitation
  if random_int > epsilon:
     # Execute a a√ß√£o com o maior valor dado um estado
     # np.argmax pode ser √∫til aqui
    action = greedy_policy(Qtable, state)
  # else --> exploration
  else:
    action = env.action_space.sample()

  return action
```

7. Definindo os hiperpar√¢metros

Os hiperpar√¢metros relacionados √† explora√ß√£o s√£o alguns dos mais importantes.

* Precisamos garantir que nosso agente explore o espa√ßo de estados o suficiente para aprender uma boa aproxima√ß√£o de valor. Para fazer isso, precisamos ter decaimento progressivo do epsilon.
* Se voc√™ diminuir o epsilon muito r√°pido (decay_rate muito alto), voc√™ corre o risco de que seu agente fique preso, j√° que seu agente n√£o explorou o espa√ßo de estado o suficiente e, portanto, n√£o pode resolver o problema.

```python
# Par√¢metros de treinamento
n_training_episodes = 10000 # Total de epis√≥dios de treinamento
learning_rate = 0.7 # Taxa de aprendizado

# Par√¢metros de avalia√ß√£o
n_eval_episodes = 100 # N√∫mero total de epis√≥dios de teste

# Par√¢metros do ambiente
env_id = "FrozenLake-v1" # Nome do ambiente
max_steps = 99 # Max passos por epis√≥dio
gamma = 0.95 # Taxa de desconto
eval_seed = [] # A semente de avalia√ß√£o do ambiente

# Par√¢metros de explora√ß√£o
max_epsilon = 1.0 # Probabilidade de explora√ß√£o no in√≠cio
min_epsilon = 0.05 # Probabilidade m√≠nima de explora√ß√£o
decay_rate = 0.0005 # Taxa de decaimento exponencial para prob de explora√ß√£o
```

8. Rotina de Treinamento

O loop de treinamento √© assim:

    Por epis√≥dio no total de epis√≥dios de treino:

    Reduza o epsilon (j√° que precisamos cada vez menos de explora√ß√£o)
    Redefinir o ambiente

        Para passo em passos de tempo m√°ximo:
            Escolha a a√ß√£o At usar a pol√≠tica gananciosa do epsilon
            Tome a a√ß√£o (a) e observe o(s) estado(s) resultante(s) e a recompensa (r)
            Atualize o valor Q Q(s,a) usando a equa√ß√£o de Bellman Q(s,a) + lr [R(s,a) + gama * max Q(s',a') - Q(s,a)]
            Se terminar, termine o epis√≥dio
            Nosso pr√≥ximo estado √© o novo estado

```python
def train(n_training_episodes, min_epsilon, max_epsilon, decay_rate, env, max_steps, Qtable):
  for episode in tqdm(range(n_training_episodes)):
    # # Reduzir epsilon (porque precisamos cada vez menos exploration)
    epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode)
    # Redefinir o ambiente
    state = env.reset()
    step = 0
    done = False

    # repete
    for step in range(max_steps):
      # Escolha a a√ß√£o At para usar a pol√≠tica gananciosa (greedy policy) do epsilon
      action = epsilon_greedy_policy(Qtable, state, epsilon)


      new_state, reward, done, info = env.step(action)

      # Update Q(s,a):= Q(s,a) + lr [R(s,a) + gamma * max Q(s',a') - Q(s,a)]
      Qtable[state][action] = Qtable[state][action] + learning_rate * (reward + gamma * np.max(Qtable[new_state]) - Qtable[state][action])

      if done:
        break

      state = new_state
  return Qtable
```

9. Treinando o agente Q-Learning 

```python
Qtable_frozenlake = train(n_training_episodes, 
                          min_epsilon, 
                          max_epsilon, 
                          decay_rate, 
                          env, 
                          max_steps, 
                          Qtable_frozenlake)
```

10. Avalia√ß√£o do M√©todo

```python
def evaluate_agent(env, max_steps, n_eval_episodes, Q, seed):
  """
   Avalie o agente para epis√≥dios ``n_eval_episodes`` e retorne recompensa m√©dia e padr√£o de recompensa.
   :param env: O ambiente de avalia√ß√£o
   :param n_eval_episodes: N√∫mero de epis√≥dios para avaliar o agente
   :param Q: A tabela Q
   :param seed: A matriz de sementes de avalia√ß√£o (para taxi-v3)
   """
  episode_rewards = []
  for episode in tqdm(range(n_eval_episodes)):
    if seed:
      state = env.reset(seed=seed[episode])
    else:
      state = env.reset()
    step = 0
    done = False
    total_rewards_ep = 0

    for step in range(max_steps):
      # Tome a a√ß√£o (√≠ndice) que tem a recompensa futura m√°xima esperada dado aquele estado
      action = greedy_policy(Q, state)
      new_state, reward, done, info = env.step(action)
      total_rewards_ep += reward

      if done:
        break
      state = new_state
    episode_rewards.append(total_rewards_ep)
  mean_reward = np.mean(episode_rewards)
  std_reward = np.std(episode_rewards)

  return mean_reward, std_reward
     
```

11. Avaliando nosso agente Q-Learning
* Normalmente, voc√™ deve ter uma recompensa m√©dia de 1,0
* O ambiente √© relativamente f√°cil j√° que o espa√ßo de estados √© muito pequeno (16). O que voc√™ pode tentar fazer √© substitu√≠-lo pela vers√£o escorregadia, que introduz estocasticidade, tornando o ambiente mais complexo.

```python
# Evaluate our Agent
mean_reward, std_reward = evaluate_agent(env, max_steps, n_eval_episodes, Qtable_frozenlake, eval_seed)
print(f"Mean_reward={mean_reward:.2f} +/- {std_reward:.2f}")
```

1. Tipo de Espa√ßo:

3. Dimens√µes do Espa√ßo:

**Componentes do Vetor de Estado:**

Para o ambiente CartPole, o vetor de estado geralmente cont√©m as seguintes componentes:

1. Posi√ß√£o do Carrinho (cart position): Quanto o carrinho se deslocou da posi√ß√£o central (pode variar entre -4.8 e 4.8).
2. Velocidade do Carrinho (cart velocity): Velocidade com a qual o carrinho est√° se movendo (n√£o h√° limites espec√≠ficos fornecidos no ambiente padr√£o, por isso valores muito grandes s√£o usados como limites).
3. √Çngulo do P√™ndulo (pole angle): √Çngulo do p√™ndulo em rela√ß√£o √† vertical (pode variar entre aproximadamente -0.418 radianos e 0.418 radianos, cerca de ¬±24 graus).

```python
# Fun√ß√£o para treinar o modelo 
def train(env, policy_model, optimizer, num_episodes=1000, gamma=0.99):
    all_returns = []

    for episode in range(num_episodes):
        state = env.reset()  # Reinicia o ambiente no in√≠cio de cada epis√≥dio
        episode_states = []
        episode_actions = []
        episode_rewards = []

        done = False
        while not done:
            action = get_action(state)  # Seleciona uma a√ß√£o com base na pol√≠tica atual
            next_state, reward, done, _ = env.step(action)  # Executa a a√ß√£o no ambiente
            episode_states.append(state)  # Armazena o estado atual
            episode_actions.append(action)  # Armazena a a√ß√£o tomada
            episode_rewards.append(reward)  # Armazena a recompensa recebida
            state = next_state  # Atualiza o estado atual

        episode_returns = compute_returns(episode_rewards, gamma)  # Calcula os retornos descontados
        all_returns.append(sum(episode_rewards))  # Armazena o total de recompensas do epis√≥dio

        episode_states = torch.tensor(episode_states, dtype=torch.float32)
        episode_actions = torch.tensor(episode_actions, dtype=torch.long)
        episode_returns = torch.tensor(episode_returns, dtype=torch.float32)

        # Clear previous gradients
        optimizer.zero_grad()

        # Forward pass: Calcula as probabilidades das a√ß√µes para cada estado no epis√≥dio
        action_probs = policy_model(episode_states)
        
        # √çndices para a√ß√µes tomadas
        action_probs_taken = action_probs.gather(1, episode_actions.view(-1, 1))  # Obtemos as probabilidades das a√ß√µes realmente tomadas
        
        # Calcula a perda (negativo do log das probabilidades ponderadas pelos retornos)
        loss = -torch.mean(torch.log(action_probs_taken) * episode_returns)

        # Backward pass: Calcula os gradientes
        loss.backward()

        # Apply gradients to update the policy
        optimizer.step()

        # Log progress
        if episode % 50 == 0:
            total_rewards = sum(episode_rewards)
            print(f"Episode {episode}: Total Reward: {total_rewards}")  # Exibe o total de recompensas a cada 50 epis√≥dios
            if total_rewards == 500:
                break

    return all_returns
```

```python
def get_action(state):
    # Ensure the state is a tensor before passing it to the model
    state = torch.tensor(state, dtype=torch.float32).unsqueeze(0)  # Add batch dimension (1, num_states)
    
    # Get action probabilities from the model
    probabilities = policy_model(state).detach().numpy().flatten()  # Convert tensor to numpy for action selection
    action = np.random.choice(num_actions, p=probabilities)  # Sample action based on probabilities
    return action
```

```python
# Executando o treinamento
num_episodes = 1000

# Initialize the policy model (you need to know the number of states and actions)
num_states = env.observation_space.shape[0]  # Size of the state space
num_actions = env.action_space.n  # Number of possible actions

policy_model = PolicyModel(num_states, num_actions)  # Your policy model (defined earlier)

# Initialize the optimizer
learning_rate = 0.001
optimizer = torch.optim.Adam(policy_model.parameters(), lr=learning_rate)

# Start the training process
returns = train(env, policy_model, optimizer, num_episodes)

# Optionally, you can print or analyze the `returns` array after training
print(f"Training completed. Total returns: {returns}")
```

O REINFORCE, apesar de ser simples e intuitivo, possui algumas limita√ß√µes:

* Alta vari√¢ncia: O gradiente de uma pol√≠tica pode ser altamente variado entre diferentes epis√≥dios, o que faz com que o algoritmo seja inst√°vel e dif√≠cil de otimizar. Isso ocorre porque as recompensas futuras podem ser muito aleat√≥rias e dif√≠ceis de prever.

* Atualiza√ß√µes grandes e ruidosas: Como as atualiza√ß√µes dos par√¢metros s√£o baseadas em trajet√≥rias completas e n√£o em estimativas mais precisas, isso leva a grandes flutua√ß√µes nos par√¢metros da pol√≠tica, o que torna o treinamento ineficiente e demorado.

**Quando o REINFORCE funciona bem:**

Apesar de ser inconstante, o REINFORCE pode ser √∫til em cen√°rios simples e controlados, mas para problemas mais complexos, com alta variabilidade nas recompensas ou grandes dimens√µes de estado, ele pode n√£o ser ideal.

Existem alguns algoritmos  avan√ßados  como:

* **Proximal  Policy  Optimization (PPO):** √©  uma  abordagem  de aprendizado  por  refor√ßo  que  otimiza  uma  pol√≠tica  parametrizada  usando uma   forma   objetiva   que   imp√µe   uma   restri√ß√£o   de   proximidade   aos par√¢metros anteriores. * **Trust Region Policy Optimization (TRPO):** √© outro algoritmo que visa  melhorar a  estabilidade das atualiza√ß√µes  de pol√≠tica  em  compara√ß√£o com m√©todos mais simples,como REINFORCE. * **Deep Deterministic Policy Gradient (DDPG):** estende m√©todos de pol√≠ticas determin√≠sticas para espa√ßos de a√ß√£o cont√≠nuos. Ele combina elementos  de  DPG  (Deterministic  Policy  Gradient)  com  redes  neurais profundas para aprendizado em ambientes complexos.

* **Actor Critic Methods:** este √© um framework que combina elementos de avalia√ß√£o  (critic)  e  a√ß√£o  (actor)  em  um  √∫nico  modelo.  O  ator  (policy network)  determina  a  a√ß√£o  a  ser  tomada,  assumindo  a  posi√ß√£o  de agente, e  o  cr√≠tico  (value  network)  avalia  a  a√ß√£o  tomada  pelo  ator. * **Soft Actor-Critic (SAC):** SAC √© um algoritmo de aprendizado por refor√ßo que  visa  maximizar  tanto  a  recompensa  esperada  quanto  a  entropia  da pol√≠tica, promovendo explora√ß√£o em ambientes de aprendizado.

Uma utiliza√ß√£o bem recente,do nosso cotidiano,de RLbaseado em pol√≠ticas tem  sido  uma  abordagem  promissora  para  aprimorar  Modelos  de  Linguagem  de Grande Escala (LLMs),como os modelos de GPT da OpenAi e Gemini do Google. M√©todos  como  o  Proximal  Policy  Optimization  (PPO)  e  o  Trust  Region  Policy Optimization  (TRPO)  s√£o  usados  para  ajustar  as  pol√≠ticas  dos  modelos,  permitindo que eles aprendam e melhorem a partir do feedback recebido. 

Esses  m√©todos  ajudam  a  otimizar  a  tomada  de  decis√µes  dos  modelos, ajustando suas respostas para maximizar uma fun√ß√£o de recompensa definida, que pode  incluir  crit√©rios  como  coer√™ncia  da  resposta,  relev√¢ncia  e  utilidade  para  o usu√°rio. ## 3. M√©todo baseado em modelo

Os m√©todos baseados em modelo (ou model-based methods) no contexto de aprendizado por refor√ßo (RL) s√£o uma abordagem em que o agente tenta aprender ou utilizar um modelo do ambiente para tomar decis√µes e melhorar seu desempenho. Em outras palavras, o agente constr√≥i ou usa um modelo do ambiente para prever o que acontecer√° se ele tomar uma determinada a√ß√£o a partir de um estado espec√≠fico. A principal vantagem desses m√©todos √© que eles permitem que o agente planeje suas a√ß√µes de forma mais informada, j√° que **ele tem uma representa√ß√£o interna de como o ambiente se comporta**.

Diferentemente  dos  m√©todos  tradicionais,  essa  abordagem  permite  que  o agente simule diferentes a√ß√µes e antecipe suas consequ√™ncias antes de implement√°-las no mundo real, otimizando suas estrat√©gias com maior efici√™ncia - aprende com o ambiente "em mem√≥ria"/simulado

Os  m√©todos  baseados  em  modelo  oferecem  uma  vantagem  significativa:  a capacidade de planejamento. Com t√©cnicas avan√ßadas, esses m√©todos permitem que o agente preveja e avalie os resultados de suas a√ß√µes futuras, acelerando o processo de aprendizado e aumentando a precis√£o das decis√µes. 

No entanto, a efic√°cia desses m√©todos depende da constru√ß√£o de um modelo preciso  do  ambiente,  um  desafio  cont√≠nuo  que  exige  adapta√ß√£o  e  atualiza√ß√£o constante.

Os principais algoritmos s√£o:

* **Dyna-Q:** combina elementos de RL baseado em modelo e sem modelo. O agente aprende uma pol√≠tica atrav√©s de intera√ß√µes diretas com o ambiente (aprendizado  sem  modelo)  e,  simultaneamente,  utiliza  um  modelo  do ambiente  para  simular  experi√™ncias  adicionais  (aprendizado baseado  em modelo). Isso acelera o processo de aprendizado ao usar simula√ß√µes para refor√ßar   o   conhecimento   obtido   das   intera√ß√µes   reais,   permitindo   um aprendizado mais r√°pido e eficiente

* **Monte Carlo Tree Search (MCTS):** √© um m√©todo de busca heur√≠stica usado para planejar a√ß√µes em espa√ßos de estados grandes e complexos, comum em  jogos. O MCTS  constr√≥i  uma  √°rvore  de  poss√≠veis  a√ß√µes  e  estados resultantes, utilizando simula√ß√µes de Monte Carlo para estimar o valor das a√ß√µes. Ele √© eficaz em ambientes nos quais as transi√ß√µes e recompensas podem ser simuladas rapidamente, como em jogos de tabuleiro, oferecendo um equil√≠brio entre explora√ß√£o e explota√ß√£o.

O AlphaZero aprende a partir de jogos simulados contra si mesmo, utilizando um modelo treinado para prever tanto a pol√≠tica quanto o valor de cada estado. Esse algoritmo demonstrou desempenho superior em jogos complexos como xadrez, Go e Shogi, aprendendo estrat√©gias de alto n√≠vel de maneira autodidata.

* **Model-Based Policy Optimization (MBPO):** √© uma abordagem que utiliza modelos din√¢micos para melhorar a efici√™ncia do aprendizado de pol√≠ticas. O MBPO envolve a gera√ß√£o de transi√ß√µes simuladas a partir do modelo do ambiente e a otimiza√ß√£o da pol√≠tica utilizando essas transi√ß√µes simuladas, al√©m das intera√ß√µes reais.
